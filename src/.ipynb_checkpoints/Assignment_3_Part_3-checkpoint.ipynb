**1. Importing the required libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
import scipy.stats
from scipy.stats import chi2
%matplotlib inline     
sns.set(color_codes=True)
```

**2. Loading data from the data file:**

```python
df = pd.read_csv("../data/train.csv")
# To display the top 5 rows 
df.head(5)   
```

**3. Checking & processing the data**

```python
df.info()
```

**The dataset has 81 columns and 1460 rows**

**Check NaN values**

```python
#Find columns have missing values
missing_data = pd.DataFrame({'total_missing': df.isnull().sum(), 'perc_missing': (df.isnull().sum()/len(df))*100})
missing_data
```

```python
#Find columns have more than 15% missing values
missing_data[missing_data['perc_missing'] > 15]
```

```python
#drop columns have more than 15% missing values
df.drop(['LotFrontage','Alley','MasVnrType','FireplaceQu','PoolQC','Fence','MiscFeature'], axis=1, inplace = True)
```

```python
missing_data = pd.DataFrame({'total_missing': df.isnull().sum(), 'perc_missing': (df.isnull().sum()/len(df))*100})
missing_data[missing_data['perc_missing'] > 15]
```

```python
#Find columns have missing values after drop
missing_data = pd.DataFrame({'total_missing': df.isnull().sum(), 'perc_missing': (df.isnull().sum()/len(df))*100})
missing_data[missing_data['total_missing'] > 0]
```

```python
#lets drop the rows having missing values:
df.dropna(axis=0, inplace = True) 
```

**4. Determine if the sale price is associated to overall quality**

```python
plt.figure(figsize=(10,5))
sns.regplot(data=df,x='OverallQual',y='SalePrice')
plt.title("Relationship between sale price and overall quality")
plt.xlabel("Overall Quality")
plt.ylabel("Sale Price")
plt.savefig('../results/reg_saleprice_overalq.png')
```

**5. Determine if the sale price is associated to lot size**

```python
plt.figure(figsize=(10,5))
sns.regplot(data=df,x='SalePrice',y='LotArea')
plt.title("Relationship between sale price and lot size")
plt.xlabel("Sale Price")
plt.ylabel("Lot size")
plt.savefig('../results/reg_saleprice_lotarea.png')
```

**6. Determine if the sale price is associated to heating quality**

```python
sns.boxplot(x="HeatingQC", y="SalePrice", data=df)
plt.title("Relationship between sale price and heating quality")
plt.xlabel("Heating Quality and Condition")
plt.ylabel("Sale Price")
plt.savefig('../results/boxplot_saleprice_heatqc.png')
```

**7. Determine if the sale price is associated to ground living area**

```python
plt.figure(figsize=(10,5))
sns.regplot(data=df,x='GrLivArea',y='SalePrice')
plt.title("Relationship between sale price and above ground living area ")
plt.xlabel("Ground Living Area")
plt.ylabel("Sale Price")
plt.savefig('../results/reg_saleprice_grlivarea.png')
```

**8. Determine if the sale price is associated to garage area**

```python
plt.figure(figsize=(10,5))
sns.regplot(x="GarageArea", y="SalePrice", data=df)
plt.title("Relationship between sale price and garage area")
plt.xlabel("Size of garage")
plt.ylabel("Sale Price")
plt.savefig('../results/reg_saleprice_garaarea.png')
```
